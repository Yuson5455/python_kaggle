```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 要使用的列名列表。如果文件不包含标题行，那么应该显式传递标题=None。
column_names=['Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhension',
            'Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
```

```python
data = pd.read_csv(r'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data',
                 names=column_names)
```

```python
data = data.replace(to_replace='?',value=np.nan) # 将缺省的值替换成标准缺失值表示
data = data.dropna(how='any') # 丢弃代右缺失值的数据，只要有一维数据有缺失就丢弃
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 随机采样25%用来测试，75%用来构建训练模型
X_train,X_test,y_train,y_test = train_test_split(data[column_names[0:9]],data[column_names[9]],test_size=0.25,random_state=33)
```

```python
y_train.value_counts() # 只有y才可以value_counts?因为y_train和y_test 是series类型?
```

```python
y_test.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
```

```python
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
lr = LogisticRegression()
sgdc = SGDClassifier()
```

```python
lr.fit(X=X_train,y=y_train)
```

```python
lr_y_predict = lr.predict(X_test)
```

```python
sgdc.fit(X=X_train,y=y_train)
```

```python
sgdc_y_predict = sgdc.predict(X=X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print("The accuracy of LogisticRegression Classifier : " + str(lr.score(X=X_test,y=y_test)))
print(classification_report(y_true=y_test,y_pred=lr_y_predict,target_names=['benign','malignant']))
```

```python
print("The accuracy of StochasticGradientDescent Classifier : " + str(sgdc.score(X=X_test,y=y_test)))
print(classification_report(y_true=y_test,y_pred=sgdc_y_predict))
```
